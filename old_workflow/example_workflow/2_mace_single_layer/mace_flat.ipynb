```python
from mace.calculators import mace_mp
from ase.constraints import UnitCellFilter
from ase.constraints import FixCartesian
import numpy as np
from ase.io import read
from ase.optimize import LBFGS
import os
```

```python
FMAX = 0.05                 # eV/Å convergence threshold
```

```python
# ----------------------------
# Core functions
# ----------------------------
calc = mace_mp(default_dtype="float64", device="cpu", head="spice_wB97M")


def optimize_cof(input, output):
    atoms = read(input)
    indices = range(len(atoms))
    con = FixCartesian(indices, mask=[False, False, True])
    atoms.set_constraint(con)
    atoms.calc = calc

    # attach a callback to print step info
    def print_step_info():
        f = atoms.get_forces()
        max_force = np.abs(f).max()
        energy = atoms.get_potential_energy()

    ucf = UnitCellFilter(atoms)
    dyn = LBFGS(ucf)
    dyn.attach(print_step_info, interval=1)
    dyn.run(fmax=FMAX)
    atoms.write(output)

def process_cifs(input_folder, output_folder):
    os.makedirs(output_folder, exist_ok=True)
    for file_name in os.listdir(input_folder):
        if file_name.endswith('.cif'):
            input_path = os.path.join(input_folder, file_name)
            output_path = os.path.join(output_folder, file_name)
            optimize_cof(input_path, output_path)
```

```python
# ----------------------------
# Run
# ----------------------------
input_folder = "input2/"
output_folder = "output"
process_cifs(input_folder, output_folder)
```
