```python
import gs as gs

raw_nodes = "raw_nodes_single"
raw_linker = "raw_linker_single"
nodes = "nodes"
linker = "linker"
mode = "single_bond"

gs.prepare_xyz(raw_nodes, nodes, mode)
gs.prepare_xyz(raw_linker, linker, mode)
```

```python
topo = "hcb"
raw_cofs_folder = "raw_cofs"

gs.build_cof_2d(topo, nodes, linker, raw_cofs_folder)
```

```python
preopt_output_folder = "cofs_preopt"

gs.opt_mace(raw_cofs_folder, preopt_output_folder)
# gs.opt_mace(input_folder=raw_cofs_folder,output_folder=preopt_output_folder,fmax=0.05,default_dtype="float64",head="spice_wB97M",device="cpu",fix_z=True)
```

```python
gs.CheckIld().run(preopt_output_folder)
```

```python
gs.ChangeIld().run(preopt_output_folder, "cofs_ild", new_z_start=3.0, new_z_end=4.5, new_z_step=0.1)
```

```python
gs.Supercell().run(preopt_output_folder, "cofs_super", supercell_size=(2, 2, 2))
```

```python
gs.IlsSerr().run(preopt_output_folder, "cofs_serr", shift_length_start=9.0, shift_length_end=10.6, shift_length_step=1.0, shift_angle=90.0)
```

```python
gs.IlsIncl().run(preopt_output_folder, "cofs_incl", incl_length_start=12.0, incl_length_end=15.8, incl_length_step=1.0, incl_angle=45.0)
```

```python
gs.SetVacuum().run(preopt_output_folder, "cofs_vac", vacuum_top=10.0)
```

```python
gs.CenterZ().run(preopt_output_folder, "cofs_centered")
```

```python
gs.RemoveLayer().run(preopt_output_folder, "cofs_rm", remove_z=0.75, remove_tol=0.02, mode="frac")
```
